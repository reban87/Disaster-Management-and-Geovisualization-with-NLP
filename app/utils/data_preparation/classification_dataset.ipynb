### Combine Disaster dataset and Non disaster dataset

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path_to_disaster_dataset = '/content/drive/MyDrive/master thesis on NLP/Colab Notebook/Dataset/disaster_data/dataset_015_to_024_bipad/Social Media Types/datasets/classification_dataset/disaster_final.csv'
path_to_non_disaster_dataset = '/content/drive/MyDrive/master thesis on NLP/Colab Notebook/Dataset/disaster_data/dataset_015_to_024_bipad/Social Media Types/datasets/classification_dataset/non_disaster_final.csv'
classification_dataset = '/content/drive/MyDrive/master thesis on NLP/Colab Notebook/Dataset/disaster_data/dataset_015_to_024_bipad/Social Media Types/datasets/classification_dataset'
```

```python
df_disaster = pd.read_csv(path_to_disaster_dataset)
df_disaster
```

```python
# consider only text and target category for data to combine
df_disaster = df_disaster[['text', 'target']]
df_disaster
```

```python
df_non_disaster = pd.read_csv(path_to_non_disaster_dataset)
df_non_disaster
```

```python
# consider only text and target category for data to combine
df_non_disaster = df_non_disaster[['text', 'target']]
df_non_disaster
```

```python
# merging two df
merged_df = pd.concat([df_non_disaster, df_disaster], ignore_index=True)
merged_df
```

```python
# shuffle the dataset
merged_df = merged_df.sample(frac=1, ignore_index=True)
merged_df
```

```python
# save the final dataset
merged_df.to_csv('/content/drive/MyDrive/master thesis on NLP/Colab Notebook/Dataset/disaster_data/dataset_015_to_024_bipad/Social Media Types/datasets/classification_dataset/classification_dataset.csv')
```
