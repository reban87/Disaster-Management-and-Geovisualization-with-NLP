#### Data Import and visualization

```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import time
from tqdm import tqdm
```

```python
# Read the dataset
incident_df=pd.read_csv(revised_datapath)

# Create New DataFrame with Title, Hazard, Incident on, Province, District, Municapality, ward.
df = pd.DataFrame().assign(title=incident_df['Title'], hazard=incident_df['Hazard'], incident_on=incident_df['Incident on'], province=incident_df['Province'], district=incident_df['District'], municipality=incident_df['Municipality'], ward=incident_df['Ward'])
df
```

```python
#Visualize Incident based upon 'Hazard type'

# Count the occurrences of each hazard in the 'hazard_counts' column
hazard_counts = df['hazard'].value_counts()


# Plot the category counts
hazard_counts.plot(kind='bar')

# Set the plot labels
plt.xlabel('hazard')
plt.ylabel('Count')
plt.title('Hazard Counts')

# Display the plot
plt.show()
```

```python
# hazard types based upon province names

# Count the occurrences of 'hazard'  based on the 'province'
hazard_counts = df.groupby('province')['hazard'].value_counts()

# Plot the 'hazard_counts' against the 'province'
hazard_counts.unstack().plot(kind='bar', figsize=(10, 6))

# Set the plot labels
plt.xlabel('Province')
plt.ylabel('Hazard')
plt.title('Hazard Counts with Province Name')

# Adjust the spacing between the bars
plt.tight_layout()

# Display the plot
plt.show()
```
