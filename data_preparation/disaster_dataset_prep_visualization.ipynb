#### Visualization Raw Datasets

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = '/content/drive/MyDrive/master thesis on NLP/Colab Notebook/Dataset/disaster_data/dataset_015_to_024_bipad/Social Media Types/datasets/combined_datasets.csv'
```

```python
df = pd.read_csv(dataset)
df
```

```python
#Visualize Incident based upon 'Hazard type'

# Count the occurrences of each hazard in the 'hazard_counts' column
def visualize_hazard_counts(df):
    # Count the occurrences of each hazard in the 'hazard' column
    hazard_counts = df['hazard'].value_counts()

    # Plot the category counts
    hazard_counts.plot(kind='bar')

    # Set the plot labels
    plt.xlabel('Hazard Types')
    plt.ylabel('Count')
    plt.title('Hazard Counts')

    # Display the plot
    plt.show()
visualize_hazard_counts(df)
```

```python
# hazard types based upon province names

# Count the occurrences of 'hazard'  based on the 'province'
def apply_filter(df):
    hazard_counts = df.groupby('province')['hazard'].value_counts()

    # Plot the 'hazard_counts' against the 'province'
    hazard_counts.unstack().plot(kind='bar', figsize=(10, 6))

    # Set the plot labels
    plt.xlabel('Province')
    plt.ylabel('Hazard')
    plt.title('Hazard Counts with Province Name')

    # Adjust the spacing between the bars
    plt.tight_layout()

    # Display the plot
    plt.show()

apply_filter(df)
```

### **Considering following hazard types**


*   Earthquake
*   Landslides
*   Flood
*   Forest Fire



```python
# Filter out above fields
filtered_df = df[df['hazard'].isin(['Earthquake', 'Forest Fire', 'Flood', 'Landslide' ])]
filtered_df
```

```python
visualize_hazard_counts(filtered_df)
```

```python
apply_filter(filtered_df)
```
