### Non disaster related datasets
target - 0 : Non disaster/ Irrelevent data

```python
df_twt = pd.read_csv(input_datapath)
df_twt
```

```python
# considering only target 0
# df_non_disaster = df_twt[['text', 'target']]
df_non_disaster = df_twt.loc[df_twt['target'] == 0]
df_non_disaster
```

```python
df_unique = df_non_disaster['keyword'].unique()
df_unique
```

```python
# # shuffle the DataFrame rows
# df_others = df_feature.sample(frac = 1)
# df_others = df_others[:1000]
# df_others
```

```python
import pandas as pd

# List of desired hazard types
desired_hazard_types = ['flood', 'earthquake', 'landslide', 'forest fire', 'forest%20fires', 'forest%20fire']

# Filter the original DataFrame to include only desired hazard types
final_df = df_non_disaster[df_non_disaster['keyword'].isin(desired_hazard_types)]

# Create a new column 'hazard_category' and assign 'others' to any hazard types not in the desired list
final_df['hazard_category'] = final_df['keyword'].apply(lambda x: x if x in desired_hazard_types else 'others')
final_df
```

```python
# Append 'others' to the DataFrame for rows where hazard types not in the desired list
others_df = df_non_disaster[~df_non_disaster['keyword'].isin(desired_hazard_types)].copy()
others_df
```

```python
# Shuffle the rows containing 'others'
others_df = others_df.sample(frac=1, random_state=42).reset_index(drop=True)

# Add 'others' to the 'hazard_category' column
others_df['hazard_category'] = 'others'
others_df
```

```python
# Concatenate final_df and others_df
final_df = pd.concat([final_df, others_df], ignore_index=True)

# Print or use the final DataFrame
final_df
```

```python
df_unique_keyword = final_df['hazard_category'].unique()
df_unique_keyword
```

```python
first_2100_rows = final_df.iloc[:2100]
first_2100_rows
```

### Disaster related datasets

```python
# read all the csv files from the folder
csv_files = [file for file in os.listdir(social_media_dataset) if file.endswith(".csv")]
dfs = [pd.read_csv(os.path.join(social_media_dataset, csv_file)) for csv_file in csv_files]

# Combine all DataFrames into a single DataFrame
combined_df = pd.concat(dfs, ignore_index=True)

# add a column name 'target' and assign the value '1'
combined_df['target'] = 1

# Rename the "Article" column to "text"
combined_df.rename(columns={"status": "text"}, inplace=True)

# select only text and target columns
combined_df = combined_df[['title', 'hazard', 'incident_on' ,'province', 'district', 'municipality', 'ward','text', 'target']]
combined_df.to_csv("/content/drive/MyDrive/master thesis on NLP/Colab Notebook/Dataset/disaster_data/dataset_015_to_024_bipad/Social Media Types/datasets/combined_datasets.csv")
```
